# Evaluación Parcial 1: Fundamentos de Deep Learning
## Sección 1: Carga y Exploración de Datos (EDA)

En esta primera sección, procedemos a importar las librerías necesarias y a descargar el conjunto de datos **CIFAR-10**. Este dataset contiene 60,000 imágenes a color de 32x32 píxeles, divididas en 10 clases (50,000 para entrenamiento y 10,000 para pruebas).

El objetivo inicial es verificar la correcta carga de los datos comprobando sus dimensiones (shape) y visualizar una muestra para confirmar que las etiquetas corresponden correctamente a las imágenes.

```python
# Importar las librerías fundamentales
import tensorflow as tf
from tensorflow.keras import datasets
import matplotlib.pyplot as plt
import numpy as np

# 1. Cargar el dataset CIFAR-10
print("Descargando/Cargando el dataset CIFAR-10...")
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# 2. Verificar las dimensiones de los datos (EDA básico)
print("\n--- Dimensiones de los datos ---")
print(f"Imágenes de entrenamiento: {x_train.shape}")
print(f"Etiquetas de entrenamiento: {y_train.shape}")
print(f"Imágenes de prueba: {x_test.shape}")
print(f"Etiquetas de prueba: {y_test.shape}")
```

```python
# Definir los nombres reales de las 10 clases de CIFAR-10 (según la documentación)
class_names = ['Avión', 'Auto', 'Pájaro', 'Gato', 'Ciervo',
               'Perro', 'Rana', 'Caballo', 'Barco', 'Camión']

# Crear una figura para visualizar las primeras 10 imágenes
plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(2, 5, i+1) # 2 filas, 5 columnas
    plt.xticks([]) # Quitar marcas del eje x
    plt.yticks([]) # Quitar marcas del eje y
    plt.grid(False)
    plt.imshow(x_train[i])
    # Las etiquetas vienen como arrays 2D (ej: [6]), extraemos el valor con [0]
    etiqueta_real = y_train[i][0]
    plt.xlabel(class_names[etiqueta_real], fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Sección 2: Preprocesamiento de Datos

Antes de construir el modelo, es imperativo preparar los datos para facilitar la convergencia matemática de la red neuronal. Para esto aplicaremos dos transformaciones:

1. **Normalización (Min-Max Scaling):** Los valores originales de los píxeles van de 0 a 255. Al dividirlos por 255.0, escalamos los datos a un rango de [0, 1]. Esto previene gradientes inestables y acelera el aprendizaje.
2. **Codificación de Etiquetas (One-Hot Encoding):** Transformamos las etiquetas de números enteros (0 al 9) a vectores categóricos de 10 posiciones. Esto es un requisito estricto para poder utilizar la función de pérdida `categorical_crossentropy` en nuestro problema de clasificación multiclase.

```python
from tensorflow.keras.utils import to_categorical

# 1. Normalización de las imágenes (dividir por 255.0)
x_train_norm = x_train.astype('float32') / 255.0
x_test_norm = x_test.astype('float32') / 255.0

# 2. One-Hot Encoding de las etiquetas
y_train_encoded = to_categorical(y_train, 10)
y_test_encoded = to_categorical(y_test, 10)

# Verificación de los cambios
print("--- Datos después del preprocesamiento ---")
print(f"Rango de píxeles: Mínimo = {x_train_norm.min()}, Máximo = {x_train_norm.max()}")
print(f"Ejemplo de etiqueta original (índice 0): {y_train[0][0]} ({class_names[y_train[0][0]]})")
print(f"Ejemplo de etiqueta codificada (One-Hot): {y_train_encoded[0]}")
```

## Sección 3: Definición de la Arquitectura del Modelo (MLP)

Para abordar este problema de clasificación, implementaremos un Perceptrón Multicapa (MLP) utilizando la API secuencial de Keras. La arquitectura diseñada consiste en las siguientes capas:

1. **Capa de Entrada (Flatten):** Dado que un MLP solo procesa vectores unidimensionales, esta capa transforma la matriz de la imagen 3D (32x32x3) en un vector plano de 3072 elementos.
2. **Capas Ocultas (Dense):** Se añaden capas densamente conectadas para que la red aprenda patrones complejos. Utilizamos la función de activación **ReLU (Rectified Linear Unit)**, ya que mitiga el problema del desvanecimiento del gradiente y acelera la convergencia.
3. **Regularización (Dropout):** Para evitar el sobreajuste (*overfitting*) en los datos de entrenamiento, incorporamos capas de Dropout que "apagan" aleatoriamente un porcentaje de las neuronas en cada época, forzando a la red a generalizar mejor.
4. **Capa de Salida (Dense):** Una capa final de 10 neuronas (una por cada clase de CIFAR-10) con la función de activación **Softmax**. Esta función convierte las salidas brutas en probabilidades, asegurando que la suma de todas sea igual a 1, permitiendo clasificar la imagen en la categoría con mayor probabilidad.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout

# Construcción del Perceptrón Multicapa (MLP)
modelo_mlp = Sequential([
    # 1. Aplanar la imagen de 32x32x3 a un vector 1D
    Flatten(input_shape=(32, 32, 3), name="Capa_Entrada_Flatten"),

    # 2. Primera capa oculta con regularización
    Dense(512, activation='relu', name="Capa_Oculta_1"),
    Dropout(0.3, name="Dropout_1"), # Apagamos el 30% de las neuronas

    # 3. Segunda capa oculta con regularización
    Dense(256, activation='relu', name="Capa_Oculta_2"),
    Dropout(0.3, name="Dropout_2"),

    # 4. Tercera capa oculta para capturar más detalles
    Dense(128, activation='relu', name="Capa_Oculta_3"),
    Dropout(0.2, name="Dropout_3"),

    # 5. Capa de Salida (10 clases, activación Softmax)
    Dense(10, activation='softmax', name="Capa_Salida")
])

# Mostrar un resumen de la arquitectura
print("\n--- Resumen de la Arquitectura del Modelo ---")
modelo_mlp.summary()
```

## Sección 4: Compilación, Hiperparámetros y Entrenamiento

Para iniciar el proceso de aprendizaje, se configuran los siguientes elementos:

1. **Optimizador:** Se utiliza **Adam**, ya que adapta la tasa de aprendizaje de forma automática durante el entrenamiento, siendo muy eficiente para problemas de visión computacional.
2. **Función de Pérdida (Loss):** Se emplea **Categorical Crossentropy**. Esta es la métrica matemática estándar cuando se enfrentan problemas de clasificación multiclase donde las etiquetas han pasado por un proceso de One-Hot Encoding.
3. **Hiperparámetros de Entrenamiento:** * **Epochs (Épocas):** Se configuran 20 épocas. Cada época es una pasada completa de todos los datos por la red.
   * **Batch Size (Tamaño de Lote):** Se define en 64. En lugar de actualizar los pesos imagen por imagen, la red procesa grupos de 64 imágenes a la vez, logrando un equilibrio entre el uso de la memoria RAM/GPU y la estabilidad del gradiente.
   * **Validación:** Se reserva un 20% de los datos de entrenamiento (`validation_split=0.2`) para validar el modelo al final de cada época y monitorear si ocurre sobreajuste.

```python
# 1. Compilación del modelo
modelo_mlp.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

# 2. Entrenamiento del modelo
print("Iniciando el entrenamiento del modelo MLP...")
historial = modelo_mlp.fit(x_train_norm, y_train_encoded,
                           epochs=20,
                           batch_size=64,
                           validation_split=0.2, # El 20% de los 50k datos será para validación
                           verbose=1)
```

## Sección 5: Evaluación, Métricas y Visualización de Resultados

Para evaluar el desempeño real de nuestro modelo MLP, analizaremos dos aspectos fundamentales:

1. **Curvas de Aprendizaje:** Graficaremos el historial de la Función de Pérdida (Loss) y la Precisión (Accuracy) tanto para el conjunto de entrenamiento como para el de validación. Esto nos permite diagnosticar el comportamiento del modelo (por ejemplo, si existe *underfitting* o *overfitting*).
2. **Métricas de Clasificación:** Utilizaremos el conjunto de pruebas (`x_test`) que el modelo nunca ha visto para generar predicciones. Con esto, construiremos un reporte detallado que incluya **Accuracy, Precision, Recall y F1-Score** por cada clase, ofreciendo una visión integral de dónde acierta y dónde se confunde la red.

```python
# 1. Gráficos de Pérdida (Loss) y Precisión (Accuracy)
plt.figure(figsize=(14, 5))

# Gráfico de Loss
plt.subplot(1, 2, 1)
plt.plot(historial.history['loss'], label='Pérdida (Train)')
plt.plot(historial.history['val_loss'], label='Pérdida (Validación)')
plt.title('Curva de Pérdida (Loss) del Modelo')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Gráfico de Accuracy
plt.subplot(1, 2, 2)
plt.plot(historial.history['accuracy'], label='Precisión (Train)')
plt.plot(historial.history['val_accuracy'], label='Precisión (Validación)')
plt.title('Curva de Precisión (Accuracy) del Modelo')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report
import numpy as np

# 1. Hacer predicciones con los datos de prueba (los que el modelo no ha visto)
predicciones_prob = modelo_mlp.predict(x_test_norm)

# 2. Convertir probabilidades a etiquetas (el índice con mayor probabilidad)
y_pred_clases = np.argmax(predicciones_prob, axis=1)

# 3. Mostrar el cuadro resumen con las métricas solicitadas
print("\n--- Reporte de Clasificación en Datos de Prueba ---")
# y_test ya es una matriz 1D con las etiquetas reales, la usamos directo
print(classification_report(y_test, y_pred_clases, target_names=class_names))
```
